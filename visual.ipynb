```python
def network(counts):
    suis = list(counts.keys())
    nodes = cons[cons['SCUI'].isin(suis)]
    nodes['Weight'] = nodes['SCUI'].map(counts)
    nodes['Scaled_Weight'] = nodes['Weight'] * 1000
    cuis = nodes['CUI'].unique()
    edges = rels[(rels['CUI1'].isin(cuis)) & (rels['CUI2'].isin(cuis)) & (rels['CUI1'] != rels['CUI2'])]
    
#     Change CUI to SCUI as node IDs
#     re_edges = []
#     for index, row in edges.iterrows():
#         cui1, cui2 = (row['CUI1'], row['CUI2'])
#         for sui1 in nodes[nodes['CUI']==cui1]['SCUI'].values:
#             for sui2 in nodes[nodes['CUI']==cui2]['SCUI'].values:
#                 if sui1 != sui2:
#                     re_edges.append([sui1, sui2, row['RELA']])

#     re_edges = pd.DataFrame(edges, columns=['Source', 'Target', 'Rela'])
    edges.columns = ['Source', 'Target', 'Rela']
    print(edges['Rela'].value_counts())
                            
    
    G = nx.from_pandas_edgelist(edges, source='Source', target='Target', edge_attr='Rela', create_using=nx.DiGraph)
    nx.set_node_attributes(G, nodes.set_index('CUI')['SCUI'].to_dict(), 'SCUI')
#     nx.set_node_attributes(G, nodes.set_index('CUI')['CUI'].to_dict(), 'CUI')
    nx.set_node_attributes(G, nodes.set_index('CUI')['STR'].to_dict(), 'Text')
    nx.set_node_attributes(G, nodes.set_index('CUI')['STY'].to_dict(), 'STY')
    nx.set_node_attributes(G, nodes.set_index('CUI')['TYPE'].to_dict(), 'Type')
    nx.set_node_attributes(G, nodes.set_index('CUI')['Weight'].to_dict(), 'Weight')
    nx.set_node_attributes(G, nodes.set_index('CUI')['Scaled_Weight'].to_dict(), 'Scaled_Weight')
    
    return G
```

```python
G = network(counts)
```

```python
kwargs = dict(width=800, height=800, xaxis=None, yaxis=None)
opts.defaults(opts.Nodes(**kwargs), opts.Graph(**kwargs))

colors = ['#000000']+hv.Cycle('Category20').values
fb_graph = hv.Graph.from_networkx(G, nx.layout.fruchterman_reingold_layout)

bundled = bundle_graph(fb_graph.opts(cmap=colors, node_size='Scaled_Weight', edge_line_width=1,
              node_line_color='gray', node_color='Type', directed=True, arrowhead_length=0.01) )
```
